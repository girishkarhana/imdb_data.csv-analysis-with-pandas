```python
import pandas as pd
```

```python
df=pd.read_csv('/content/imdb_data.csv',on_bad_lines='skip')
print(df)
```

```python
df.head()
```

```python
df['profit']=df['revenue']-df['budget']
df['roi'] = df['profit']/df['budget']
```

```python
profitable_movies = df.loc[df['profit'].idxmax(), ['original_title','belongs_to_collection','cast','production_companies','profit']]
print(profitable_movies)
```

```python
language_roi=df.loc[df['roi'].idxmax(),['spoken_languages','roi']]
print(language_roi)
```

```python
unique_genres = df['genres'].unique()
print(unique_genres)
```

```python
producer_director = df[['production_companies']]
print(producer_director)
p_roi=df.loc[df['roi'].idxmax(),['production_companies','roi']]

top_3_producers= df.nlargest(3, 'roi')[['production_companies', 'roi']]

print(top_3_producers)
```

```python
def extract_actors_pandas(cast_column):
    actor_list = []
    for entry in cast_column.dropna():
        try:
            cast_data = eval(entry)
            actor_list.extend([actor['name'] for actor in cast_data if 'name' in actor])
        except (ValueError, SyntaxError, TypeError):
            continue
    return actor_list

all_actors_pandas = extract_actors_pandas(df['cast'])
actor_counts_df = pd.DataFrame({'actor': all_actors_pandas})
actor_counts_summary = actor_counts_df['actor'].value_counts()
most_movies_actor_pandas = actor_counts_summary.idxmax(), actor_counts_summary.max()

most_movies_actor_pandas
```

```python
most_repeated_value = df['cast'].value_counts().head(2)
most_repeated_count = df['cast'].value_counts().max()

most_rep=most_repeated_value.iloc[1]
print(most_rep)
print(f"Number of repetitions: {most_repeated_count}")
```
